## **RANDOM FOREST**

```python
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import time

# Random Forest
def rf_cl_bo(max_depth, max_features, n_estimators, min_samples_split, min_samples_leaf):
    params_rf = {}
    params_rf['max_depth'] = round(max_depth)
    params_rf['max_features'] = max_features
    params_rf['n_estimators'] = round(n_estimators)
    params_rf['min_samples_split'] = round(min_samples_split)
    params_rf['min_samples_leaf'] = round(min_samples_leaf)
    scores = cross_val_score(RandomForestClassifier(random_state=123, **params_rf),
                             x_train, y_train, cv=5).mean()
    return scores

# Run Bayesian Optimization for Random Forest
start = time.time()
params_rf = {
    'max_depth': (3, 10),
    'max_features': (0.1, 1.0),
    'n_estimators': (80, 150),
    'min_samples_split': (2, 20),
    'min_samples_leaf': (1, 10)
}
rf_bo = BayesianOptimization(rf_cl_bo, params_rf, random_state=111)
rf_bo.maximize(init_points=20, n_iter=4)
print('It takes %s minutes' % ((time.time() - start)/60))

best_params_rf = rf_bo.max['params']
best_params_rf['max_depth'] = round(best_params_rf['max_depth'])
best_params_rf['n_estimators'] = round(best_params_rf['n_estimators'])

print("Best Random Forest Parameters:", best_params_rf)
```

## **XGBOOST**

```python
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
import numpy as np
import time

# XGBoost
def xgb_cl_bo(max_depth, learning_rate, n_estimators, min_child_weight, subsample, colsample_bytree):
    params_xgb = {}
    params_xgb['max_depth'] = int(round(max_depth))
    params_xgb['learning_rate'] = learning_rate
    params_xgb['n_estimators'] = int(round(n_estimators))
    params_xgb['min_child_weight'] = int(round(min_child_weight))
    params_xgb['subsample'] = subsample
    params_xgb['colsample_bytree'] = colsample_bytree
    scores = cross_val_score(XGBClassifier(random_state=123, **params_xgb),
                             x_train, y_train, cv=5).mean()
    return scores

# Run Bayesian Optimization for XGBoost
start = time.time()
params_xgb = {
    'max_depth': (3, 10),
    'learning_rate': (0.01, 1.0),
    'n_estimators': (80, 150),
    'min_child_weight': (1, 10),
    'subsample': (0.8, 1.0),
    'colsample_bytree': (0.8, 1.0)
}
xgb_bo = BayesianOptimization(xgb_cl_bo, params_xgb, random_state=111)
xgb_bo.maximize(init_points=20, n_iter=4)
print('It takes %s minutes' % ((time.time() - start)/60))

best_params_xgb = xgb_bo.max['params']
best_params_xgb['max_depth'] = int(round(best_params_xgb['max_depth']))
best_params_xgb['n_estimators'] = int(round(best_params_xgb['n_estimators']))
best_params_xgb['min_child_weight'] = int(round(best_params_xgb['min_child_weight']))

print("Best XGBoost Parameters:", best_params_xgb)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from hyperopt import hp, fmin, tpe, Trials

# Standardize features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Define the search space for hyperparameters
space = {
    'num_hidden_layers': hp.choice('num_hidden_layers', [1, 2, 3]),
    'num_units': hp.choice('num_units', [32, 64, 128]),
    'activation': hp.choice('activation', ['relu', 'sigmoid']),
    'optimizer': hp.choice('optimizer', ['adam', 'sgd']),
    'batch_size': hp.choice('batch_size', [32, 64, 128]),
    'epochs': hp.choice('epochs', [10, 20, 30])
}

# Define the objective function for hyperparameter optimization
def objective(params):
    model = Sequential()
    model.add(Dense(units=params['num_units'], activation=params['activation'], input_shape=(x_train.shape[1],)))
    for _ in range(params['num_hidden_layers'] - 1):
        model.add(Dense(units=params['num_units'], activation=params['activation']))
    model.add(Dense(units=1, activation='sigmoid'))

    model.compile(optimizer=params['optimizer'], loss='binary_crossentropy', metrics=['accuracy'])

    model.fit(x_train_scaled, y_train, batch_size=params['batch_size'], epochs=params['epochs'], verbose=0)

    y_pred_proba = model.predict(x_test_scaled)
    y_pred = (y_pred_proba > 0.5).astype(int)  # Convert probabilities to classes

    acc = accuracy_score(y_test, y_pred)

    return -acc  # minimize negative accuracy

# Perform Bayesian optimization
trials = Trials()
best = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=10, trials=trials)

print("Best hyperparameters:", best)
```
